```python
import requests
import pandas as pd 
```

```python
# pull in the api data using the requests module and assign it to 'r'
r = requests.get("http://universities.hipolabs.com/search?country=United+States")

# determine the type of data I have.
r.headers['content-type']
```

```python
# I've determined that I have a json so let's pull it into a dataframe.
data= r.json()
df= pd.DataFrame(data=data)
df
```
