```python
import pandas as pd
import matplotlib.pyplot as plt
```

United States Carbon Emissions

```python
df = pd.read_csv('assets/co2_emissions_kt_by_country.csv')
df.head()
```

```python
df.rename(columns= {'value': 'carbon_emissions'}, inplace= True)
```

```python
ndf= df.drop(columns= 'country_code')
ndf2 = ndf[ndf['country_name'].str.contains('United States')]
ndf2.head()
```

```python
ndf2.plot.line(x ='year', rot= 45, y= 'carbon_emissions', title = 'United States Carbon Emissions', figsize= (10, 5), grid= True)
```

```python
fig, ax = plt.subplots()
ax.plot(ndf2['year'], ndf2['carbon_emissions'])
ax.set_xlabel('Year')
ax.set_ylabel('Emissions  (in kilotons)')
ax.set_title("United States Carbon Emissions")
ax.ticklabel_format(style='plain')
plt.show()

```
